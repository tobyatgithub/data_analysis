### 3. Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
weather_data['Date'] = pd.to_datetime(weather_data.Date,infer_datetime_format=True)
weather_data['MONTH'] = weather_data["Date"].apply(lambda x: x.month)
weather_data.head()
# weather_data.describe()
# weather_data.Date.apply()
```

```python
weather_data.columns
```

```python
print("Here's the mean precipitation each month.")
weather_data.groupby(["MONTH"]).agg({
    "Precipitation_In":"mean",
#     "Precipitation_In":"median",
})
```

```python
print("Here's the median precipitation each month.")
weather_data.groupby(["MONTH"]).agg({
#     "Precipitation_In":"mean",
    "Precipitation_In":"median",
})
```

### 4. What’s the average number of bikes at a given bike station?

```python
station_data.head()
```

```python
ave_bike = station_data.groupby("station_id").agg({"current_dockcount": "mean"}).mean()
print("There's %0.2f bikes on average at any given bike station" % round(ave_bike,2))
```

### 6. What's the month when we have most bike station installed?

```python
station_data['install_date'] = pd.to_datetime(station_data['install_date'],infer_datetime_format=True)
station_data["month"] = station_data['install_date'].apply(lambda x: x.month)
station_data.head()
```
