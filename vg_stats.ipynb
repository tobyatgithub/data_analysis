### 9. Culture difference. Are customers from NA, EU, and JP different from each other?

```python
tmp = vs_data[['Genre',"NA_Sales","EU_Sales","JP_Sales"]].groupby("Genre").agg({
    "NA_Sales":"sum",
    "EU_Sales":"sum",
    "JP_Sales":"sum"
})
tmp
```

```python
# to compare it, we normalize each Sales data columnms, and then scale it

tmp["NA_Sales"] = (tmp["NA_Sales"]-tmp["NA_Sales"].mean()) / tmp["NA_Sales"].std()
tmp["EU_Sales"] = (tmp["EU_Sales"]-tmp["EU_Sales"].mean()) / tmp["EU_Sales"].std()
tmp["JP_Sales"] = (tmp["JP_Sales"]-tmp["JP_Sales"].mean()) / tmp["JP_Sales"].std()

tmp["NA_Sales"] = tmp["NA_Sales"] / tmp["NA_Sales"].max()
tmp["EU_Sales"] = tmp["EU_Sales"] / tmp["EU_Sales"].max()
tmp["JP_Sales"] = tmp["JP_Sales"] / tmp["JP_Sales"].max()

# alright..divide by max looks more intutitve here

print("Here we can see that NA players are very similar to EU players\n\
The only tiny two differences are \n1. NA players prefer platform games more than EU ones \n\
2. EU players like racing games more than NA ones.\n\n\
On the other hand, JP players are very different. They show nomral interest to the \n\
best sales genre in EU and NA (action). And heavily prefer Role-playing.")
tmp
```
